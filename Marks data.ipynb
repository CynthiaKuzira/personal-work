DATA CLEANING IN PYTHON

```python
# Importing the packages
import pandas as pd
```

```python
#Loading the dataset
Marks_data = pd.read_excel('Marks_data.xlsx')
Marks_data
```

FILL IN MISSING VALUES

```python
# Filling the Nan values with "unknown"
cleaned_data =Marks_data.fillna('unknown')
cleaned_data
```

DROPPING THE Nan Values

```python
#Dropping the Nan Values
cleaned_data = Marks_data.dropna()
cleaned_data
```

DROP THE DUPLICATE VALUES

```python
#find duplicate rows across all columns
duplicateRows = Marks_data[Marks_data.duplicated()]
duplicateRows
```

```python
cleaned_data =Marks_data.drop_duplicates()
cleaned_data
```

FIX INCONSISTENT LABELS

```python
# Getting the list of distinct subjects
subjects =Marks_data.Subject.dropna().unique()
sorted(subjects)
```

```python
# Creating a copy of the cleaned data
cleaned_data =Marks_data.copy(deep=True)
```

```python
# fixing the inconsistent labels
cleaned_data.update(pd.Series(['Math'],name='Subject',index=[5]))
cleaned_data
```

```python
# fixing letters withnin numbers
cleaned_data.update(pd.Series([70],name='Marks', index=[3]))
cleaned_data
```
